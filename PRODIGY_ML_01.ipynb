```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_data = pd.read_csv("housing.csv")
```

```python
housing_data
```

```python
housing_data.info
```

```python
housing_data.dropna(inplace = True)
```

```python
housing_data.info()
```

```python
from sklearn.model_selection import train_test_split

x = housing_data.drop(['median_house_value'], axis = 1)
y = housing_data['median_house_value']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
train_housing_data = x_train.join(y_train)
```

```python
train_housing_data
```

```python
train_housing_data.hist(figsize = (13, 9))
```

```python
plt.figure(figsize = (9, 6))
sns.heatmap(housing_data.corr(numeric_only = True), annot = True, cmap = "YlOrRd")
plt.show()
```

```python
train_housing_data['total_rooms'] = np.log(train_housing_data['total_rooms'] + 1)
train_housing_data['total_bedrooms'] = np.log(train_housing_data['total_bedrooms'] + 1)
train_housing_data['population'] = np.log(train_housing_data['population'] + 1)
train_housing_data['households'] = np.log(train_housing_data['households'] + 1)
```

```python
train_housing_data.hist(figsize = (13, 9))
```

```python
train_housing_data = train_housing_data.join(pd.get_dummies(train_housing_data.ocean_proximity)).drop(['ocean_proximity'], axis = 1)
```

```python
train_housing_data
```

```python
plt.figure(figsize = (14, 7))
sns.heatmap(train_housing_data.corr(), annot = True, cmap = "YlOrRd")
plt.show()
```

```python
plt.figure(figsize = (10, 7))
sns.scatterplot(x = "latitude", y = "longitude", data = train_housing_data, hue = "median_house_value")

# we can see here that the more darker it gets the more expensive the houses and the lighter the color gets the less expensive the houses in the figure so all houses that in the coast are expensive which is represented by the dark color and for lighter colors it's the inland houses
```

```python
train_housing_data['bedroom_ratio'] = train_housing_data['total_bedrooms'] / train_housing_data['total_rooms']
train_housing_data['household_rooms'] = train_housing_data['total_rooms'] / train_housing_data['households']
```

```python
plt.figure(figsize = (15, 7))
sns.heatmap(train_housing_data.corr(), annot = True, cmap = "YlOrRd")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train, y_train = train_housing_data.drop(['median_house_value'], axis = 1), train_housing_data['median_house_value']
x_train_s = scaler.fit_transform(x_train)

model = LinearRegression()
model.fit(x_train_s, y_train)
```

```python
test_housing_data = x_test.join(y_test)

test_housing_data['total_rooms'] = np.log(test_housing_data['total_rooms'] + 1)
test_housing_data['total_bedrooms'] = np.log(test_housing_data['total_bedrooms'] + 1)
test_housing_data['population'] = np.log(test_housing_data['population'] + 1)
test_housing_data['households'] = np.log(test_housing_data['households'] + 1)

test_housing_data['bedroom_ratio'] = test_housing_data['total_bedrooms'] / test_housing_data['total_rooms']
test_housing_data['household_rooms'] = test_housing_data['total_rooms'] / test_housing_data['households']
```

```python
x_test, y_test = test_housing_data.drop(['median_house_value'], axis = 1), test_housing_data['median_house_value']
```

```python
x_test_s = scaler.transform(x_test)
```

```python
model.score(x_test_s, y_test)
```

```python
# Make predictions on the test set
predictions = model.predict(x_test_s)
```

```python
# Evaluate the model
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predictions)
print(f'Mean Squared Error: {mse}')
```

```python
# Plot predictions against actual values
plt.scatter(y_test, predictions)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Linear Regression Model Performance')
plt.show()
```
